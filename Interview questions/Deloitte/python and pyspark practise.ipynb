```python
# if a number has appered atleast 3 time and once it should be concurrent
a = [10,20,20,30,40,50,20,30,40,50,70,80,90]
def occ(l,target):
    occ=0
    for i in l:
        if i == target:
            occ += 1
        
    return occ

def isConcurrent(l,target):
    n=len(l)
    for i in range(n-1):
        if l[i] == l[i+1]:
            return True

set_a = set(a)
for i in set_a:
    if occ(a,i)>=3 and isConcurrent(a,i):
        print(i)
```
