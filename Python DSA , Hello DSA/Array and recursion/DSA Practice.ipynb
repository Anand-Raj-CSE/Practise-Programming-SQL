```python
def two_sum_dict(nums,target):
    my_dict = {}
    my_ans = []
    for i in nums : 
        if target-i in nums and target-i not in my_dict:
            my_dict[i] = target-i
            my_ans.append([i,target-i])
        else : 
            my_dict[i] = target-i
    return my_ans
```

```python
target = 9
print(two_sum_dict([5,2,4,8,1,7,2,3,4,5],9))
```

The above code of n^2 complexity as it is iterating over the array twice , below am trying of better code with lesser complexity.

```python
def two_sum_opt(nums,target):
    my_dict = {}
    ans = []
    for idx,num in enumerate(nums):
        remaining = target - num
        if remaining in my_dict:
            ans.append([[idx,num],[my_dict[remaining],remaining]])
        my_dict[num] = idx
    return ans 
        
```

```python
target = 9
print(two_sum_opt([5,2,4,8,1,7,2,3,4,5],9))
```

Container with most water : https://leetcode.com/problems/container-with-most-water/submissions/1871819419/

```python
def max_area_tap(height):
    left , right = 0 , len(height)-1
    max_area = 0
    while(left<=right):
        width = right - left
        length = min(height[left],height[right])
        curr_area = width * length
        max_area = max(curr_area,max_area)
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1
    return max_area
```

rotate matrix 90 degress clockwise

```python
def rotate_matrix(matrix):
    n = len(matrix)
    #Transposing the matrix
    for i in range(n):
        for j in range(i+1,n):
            matrix[i][j] , matrix[j][i]=matrix[j][i],matrix[i][j]
    
    for i in matrix:
        i.reverse()
```

```python
l = [
  [1,2,3],
  [4,5,6],
  [7,8,9]
 ]
rotate_matrix(l)
print(l)
```
