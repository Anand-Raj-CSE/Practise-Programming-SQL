```python
# remove a particular character
def remove_ch(s,ch):
    if len(s)==0:
        return s
    small_ans = remove_ch(s[1:],ch)
    if s[0] == ch :
        return small_ans
    else:
        return s[0]+small_ans
```

```python
s='good morningzzzzz'
print(remove_ch(s,'z'))
```

# Checking palindrome using recursion

```python
def palindrome(s):
    n=len(s)
    if n<=1:
        return True
    if s[0] == s[n-1] and palindrome(s[1:n-1]):
        return True
    return False

print(palindrome('aba'))
```

```python
print(palindrome('abab'))
print(palindrome('ababa'))
print(palindrome('NItin'))
```
