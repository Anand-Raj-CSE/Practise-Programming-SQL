# substring - order should be maintained suppose we have abc --> a,b,c,ab,bc
## We could be considering the fact that they should be continuous , we could not skip like ac , and also reverse is not true as in ca

# Subsequence : Here order is maintained but they could be non continuous as well. so for abc we have subsequence of a,b,c,ab,bc,ca,[] , [] could also be a subsequence.

### For substring we could get (n*(n+1))/2 , but for subsequence we have 2^n

```python
def subsequence(s,l=[]):
    if len(s)<=1:
        return s
    l.append(s[0])
    rest_substr = subsequence(s[1:],l)
    l.append()
    return l
```

Correct approach and code

```python
def get_subsequence(s):
    if s == '':
        return ['']
    small_ans = get_subsequence(s[1:])
    my_char = s[0]
    ans = []
    ans.extend(small_ans)
    for each_subseq in small_ans:
        ans.append(my_char+each_subseq)
    return ans
```

```python
print(get_subsequence('abc'))
```

Leetcode similar question : https://leetcode.com/problems/subsets/description/

```python
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        if nums == []:
            return [[]]
        small_ans = self.subsets(nums[1:])
        ans = []
        ans.extend(small_ans)
        my_num = nums[0]
        for each_permutation in small_ans:
            ans.append([my_num]+each_permutation)
        return ans
```
