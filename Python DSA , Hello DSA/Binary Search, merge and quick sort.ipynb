```python
def BinarySearchHelper(l,x,s,e):
    if s<=e:
        mid = s+(e-s)//2
        if l[mid] == x:
            return mid
        if x<l[mid]:
            return BinarySearchHelper(l,x,s,mid-1)
        return BinarySearchHelper(l,x,mid+1,e)
    else:
        return -1
```

```python
def binarySearch(l,x):
    return BinarySearchHelper(l,x,0,len(l))

l = [1,2,3,4,5,7,8,9,10,13,45,85,779]
print(binarySearch(l,45))
```

```python
def Merge(l,s,m,e):
  # Two pointer , one for right part of array and another for left part of array which would hepl us in 
  i=s
  j = m+1
  ans = []
  while(i<=m and j<=e):
    if (l[i]<l[j]):
      ans.append(l[i])
      i += 1
    elif(l[j]<l[i]):
      ans.append(l[j])
      j += 1
    elif(l[j] == l[i]):
      ans.append(l[i])
      ans.append(l[j])
      i+=1
      j+=1
  while(i<=m):
    ans.append(l[i])
    i+=1
  while(j<=e):
    ans.append(l[j])
    j+=1
  # We have to update our master list as well
  startOf_mainList = 0
  startOfOurList = s
  while(startOfOurList<=e):
    l[startOfOurList] = ans[startOf_mainList]
    startOf_mainList += 1
    startOfOurList += 1
  return 
```

```python
def MergeSort(l,s,e):
  if s>=e:
    return
  mid = s+(e-s)//2
  MergeSort(l,s,mid)
  MergeSort(l,mid+1,e)
  Merge(l,s,mid,e)
```

```python
l = [10,2,54,12,69,85,14]
MergeSort(l,0,len(l)-1)
print(l)
```

```python
def PartitionFunction(l1,s,e):
  i=s
  rp=s
  pivot = l1[e]
  for i in range(s,e):
    if l1[i]<pivot:
      rp+=1
    i+=1
  l1[e],l1[rp] = l1[rp],pivot
  pivotIndex = rp
  #now we have to make sure that everthing in right of pivot index is greater than our pivot element and in left we have smaller 
  start = s
  end = e
  while(start<pivotIndex and end > pivotIndex):
    if(l1[start]<pivot):
      start +=1
    elif(l1[end]>=pivot):
      end -= 1
    else:
      l1[start],l1[end] = l1[end],l1[start]
      start += 1
      end -= 1
  return pivotIndex
```

```python
def quick_sort(l1,s,e):
  if s>=e:
    return
  pivot = PartitionFunction(l1,s,e)
  quick_sort(l1,s,pivot-1)
  quick_sort(l1,pivot+1,e)
  return
```

```python
l = [3,6,7,2,1,4,5]
quick_sort(l,0,len(l)-1)
print(l)
```
