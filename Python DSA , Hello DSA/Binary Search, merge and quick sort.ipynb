```python
def BinarySearchHelper(l,x,s,e):
    if s<=e:
        mid = s+(e-s)//2
        if l[mid] == x:
            return mid
        if x<l[mid]:
            return BinarySearchHelper(l,x,s,mid-1)
        return BinarySearchHelper(l,x,mid+1,e)
    else:
        return -1
```

```python
def binarySearch(l,x):
    return BinarySearchHelper(l,x,0,len(l))

l = [1,2,3,4,5,7,8,9,10,13,45,85,779]
print(binarySearch(l,45))
```

```python
def Merge(l,s,m,e):
  # Two pointer , one for right part of array and another for left part of array which would hepl us in 
  i=s
  j = m+1
  ans = []
  while(i<=m and j<=e):
    if (l[i]<l[j]):
      ans.append(l[i])
      i += 1
    elif(l[j]<l[i]):
      ans.append(l[j])
      j += 1
    elif(l[j] == l[i]):
      ans.append(l[i])
      ans.append(l[j])
      i+=1
      j+=1
  while(i<=m):
    ans.append(l[i])
    i+=1
  while(j<=e):
    ans.append(l[j])
    j+=1
  # We have to update our master list as well
  startOf_mainList = 0
  startOfOurList = s
  while(startOfOurList<=e):
    l[startOfOurList] = ans[startOf_mainList]
    startOf_mainList += 1
    startOfOurList += 1
  return 
```

```python
def MergeSort(l,s,e):
  if s>=e:
    return
  mid = s+(e-s)//2
  MergeSort(l,s,mid)
  MergeSort(l,mid+1,e)
  Merge(l,s,mid,e)
```

```python
l = [10,2,54,12,69,85,14]
MergeSort(l,0,len(l)-1)
print(l)
```
